```python
input_folder = r'\\10.99.68.178\andreex\data\monkey fetus\gestational 40'
output_folder = r'\\10.99.68.178\andreex\data\monkey fetus\gestational 40\segmentation_analysis\pix_res_info'
os.makedirs(output_folder, exist_ok=True)
```

```python
ndpi_files = sorted([file for file in os.listdir(input_folder) if file.endswith('.ndpi')])
```

```python
ndpi_files
```

```python
for filename in ndpi_files:
    path = os.path.join(input_folder, filename)
    print(path)
    slide = openslide.OpenSlide(path)
    #for i in slide.properties:
    #    print(f'{i}: {slide.properties[i]}')

    pix_res = {'x': slide.properties['openslide.mpp-x'],
               'y': slide.properties['openslide.mpp-y']}

    print(pix_res)
    #pix_res = {'x': slide.properties['openslide.PROPERTY_NAME_BOUNDS_X'],
    #           'y': slide.properties['openslide.PROPERTY_NAME_BOUNDS_Y']}
    mat_path = os.path.join(output_folder, f"{(os.path.splitext(filename)[0] + '.mat')}")
    sio.savemat(mat_path, {'pix_res': pix_res})

print('Done extracting pixel sizes and saving to output folder.')
```

```python
# get indices of organs:
indices_list = []

cur = []
for i in class_labels:
    if i == 0:
        pass
    elif len(cur) > 0:
        if cur == [15]:
            indices_list.append(cur)
            cur = [i]
        elif int(cur[-1]) == int(i - 1):
            cur.append(i)
        else:
            indices_list.append(cur)
            cur = [i]
    else:
        cur.append(i)

indices_list.append(cur)
print(indices_list)

#classes = noise + adrenal + thymus + thyroid + bone + brain + ear + ecm + eye + gi + gonads + heart + kidney + liver + lungs + mesokidney
```

```python
organ_grouped_density_dict = {}
organ_grouped_cell_count_dict = {}
organ_grouped_volume_dict = {}


for org_num, sublist in enumerate(indices_list):
    #print(sublist)
    vol = 0
    cell_count = 0 
    for i in sublist:
        vol += cell_volume_count[i]
        cell_count += cell_count_dict[i]
    
    organ_grouped_cell_count_dict[org_num] = cell_count
    organ_grouped_volume_dict[org_num] = vol
    
    #print(cell_count/vol)
    
    organ_grouped_density_dict[org_num] = cell_count/vol
```

```python
noise_labels = []

for index, element in enumerate(classes):
    if 'noise' in element.lower() or 'non_tissue' in element.lower() or 'nontissue' in element.lower():
        noise_labels.append(index)
        #print(f"'noise' found at index {index} in the list.")

print(noise_labels)
```

```python
# exclude noise

import scipy.stats as stats

log_cell_count_dict = {key: math.log(value) for key, value in cell_count_dict.items()}
log_cell_volume_count = {key: math.log(value) for key, value in cell_volume_count.items()}

# plot each label individually

y_values = list(log_cell_count_dict.values())[1:] # exclude noise class outlier
x_values = list(log_cell_volume_count.values())[1:]# exclude noise class outlier

x_values = [f for i,f in enumerate(x_values) if i not in noise_labels]
y_values = [f for i,f in enumerate(y_values) if i not in noise_labels]

corr, p_value = stats.pearsonr(y_values, x_values)
m, b = np.polyfit(y_values, x_values, 1)

xfit = np.linspace(min(y_values), max(y_values), num=5)
yfit = m*xfit + b


plt.figure(figsize=(18, 6))

plt.plot(yfit, xfit, '--', color='black', linewidth=3)
plt.scatter(x_values, y_values)
plt.xlabel('log Volume')
plt.ylabel('log Cell Count')
plt.title('Cell Count vs Volume for each label')

plt.text(0.1, 0.85, f'Correlation: {corr:.2f}\nP-value: {p_value:.2e}\nm: {m:.2f}\nb: {b:.2f}', transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

```python
# %matplotlib notebook

fig, ax1 = plt.subplots(figsize=(12,6))

y_values = list(cell_count_dict.values())[1:] # exclude noise class outlier
x_values = list(cell_volume_count.values())[1:]# exclude noise class outlier

x_values = [f for i,f in enumerate(x_values) if i not in noise_labels]
y_values = [f for i,f in enumerate(y_values) if i not in noise_labels]

ax1.scatter(x_values, y_values)

ax1.set_xscale('log')
ax1.set_yscale('log')

ax1.set_title('Cell Count vs Volume for Rhesus Macaque Fetus Organs')
ax1.set_xlabel('Volume (pixel count)')
ax1.set_ylabel('Cell Count')

# Line of best fit
log_vols = np.log10(x_values)
log_counts = np.log10(y_values)

m, b = np.polyfit(log_vols, log_counts, 1) 

corr, p_value = stats.pearsonr(log_vols, log_counts)

plt.text(0.1, 0.85, f'Correlation: {corr:.2f}\nP-value: {p_value:.2e}\nm: {m:.2f}\nb: {b:.2f}', transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.5))

xfit = np.linspace(np.min(log_vols), np.max(log_vols), num=5)
yfit = m*xfit + b

ax1.plot(10**xfit, 10**yfit, '--', color='black', linewidth=3)

plt.show()
```

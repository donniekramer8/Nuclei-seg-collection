```python
import json
import os
import numpy as np
```

```python
pth_json = r'\\10.162.80.16\Andre_expansion\data\monkey_fetus\Stardist\StarDist_12_25_23\json'
json_list = [os.path.join(pth_json, f) for f in os.listdir(pth_json)]
```

```python
for json_pth in json_list:
    print(json_pth)
    data = json.load(open(json_pth))
    break
```

```python
len(data)
```

```python
json_data = []

old_centroids = []
new_centroids = []

for cell in data:
    centroid_old = cell['centroid'][0]
    contour = cell['contour'][0]
    
    #print(centroid_old)
    #print(contour)
    
    centroid_new = [round(np.mean(contour[0])),round(np.mean(contour[1]))]
    #print([int(centroid_new[0]),int(centroid_new[1])])
    #print(centroid_new)
    # Create a new dictionary for each contour
    dict_data = {
        "centroid": [centroid_new],
        "contour": [contour]
    }
    
    # testing, delete
    old_centroids.append(centroid_old)
    new_centroids.append(centroid_new)

    json_data.append(dict_data)
```

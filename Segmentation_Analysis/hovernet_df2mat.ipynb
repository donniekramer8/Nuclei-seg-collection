```python
import pandas as pd
import numpy as np
from scipy.io import savemat
import os
```

```python
#src = r"\\10.99.68.178\andreex\data\monkey fetus\gestational 40\hovernet_out\df"
#dfs = sorted([_ for _ in os.listdir(src) if _.endswith('pkl')])

src = r'\\10.99.68.178\andreex\data\monkey fetus\gestational 40\segmentation_analysis\model1_class_info_pickles_12_4'
dfs = sorted([_ for _ in os.listdir(src) if _.endswith('.pickle')])
```

```python
print(dfs)
```

```python
for dfnm in dfs[::-1]:
    print("Saving: {}".format(dfnm))
    dst = os.path.join(*[os.path.dirname(src),'mat',dfnm.replace('pkl','mat')])
    if os.path.exists(dst):
        print("pkl already exists, skipping the file ID {}".format(dfnm))
        continue

    df = pd.read_pickle(os.path.join(src,dfnm))
    df = df[['Centroid_x', 'Centroid_y']]
    df = [_ for _ in df.to_numpy()]
    df = np.array(df)
    print(dst)
    
    savemat(dst, {'centroid':df})
```
